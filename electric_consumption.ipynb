```python
def data_info(water):
    cols = []
    unique = []
    n_uniques = []
    dtypes = []
    nulls = []

    for col in water.columns:
        cols.append(col)
        dtypes.append(water[col].dtype)
        n_uniques.append(water[col].nunique())
        unique.append(water[col].unique())
        nulls.append(water[col].isna().sum())

    return pd.DataFrame({'Columns': cols, 'n_uniques': n_uniques,
                         'unique': unique, 'dtypes': dtypes, "NULLS": nulls
                         })
```

```python

site = buildingname.split("_")[0]
weather_data_site = weather[weather.site_id == site].truncate(before='2017-01-01')
```

```python
weather_data_site.info()
```

```python
# Step 1: Resample weather data to hourly and take the average
# This ensures that every hour has one consistent reading per feature
weather_data_site_clean = weather_data_site.drop(columns=["site_id"])
weather_hourly = weather_data_site_clean.resample("H").mean()

# Step 2: Remove outliers below -40 (very likely faulty sensor readings)
# Applies to all columns â€” removes any value below -40 across the DataFrame
weather_hourly_nooutlier = weather_hourly[weather_hourly > -40]

# Step 3: Fill any resulting gaps using forward-fill
# This keeps the time series continuous by copying the last valid value forward
weather_hourly_nooutlier_nogaps = weather_hourly_nooutlier.fillna(method='ffill')
```

```python
temperature = weather_hourly_nooutlier_nogaps["airTemperature"]
```

```python
temperature.plot()
```

```python
# Step 1: Combine electricity, water, and temperature into a single DataFrame
# This aligns all three data sources based on timestamp (index)
df = pd.concat([elec_series_prediction_data, water_series_prediction_data, temperature], axis=1)

# Step 2: Rename the columns for clarity
# These will be used as features and the prediction target
df.columns = ["electricity_kW", "water_m3", "temperature_C"]

# Step 3: Drop any rows that contain missing values in any of the three columns
# This ensures we have clean, aligned data for model training
df = df.dropna()

# Step 4: Add a lag feature from water usage (previous hour's water consumption)
# This helps the model learn from recent past behavior, which often influences current electricity usage
df["water_m3_lag1"] = df["water_m3"].shift(1)

# Step 5: Drop the first row (and any others) that now contain NaN due to the lag shift
# This ensures the dataset is clean and ready for modeling
df = df.dropna()
```

```python
df
```

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
import pandas as pd

# Define months for train/test
training_months = [4, 5, 6]
test_months = [7]

# --- STEP 1: Prepare features (match notebook format) ---

# Extract hour and dayofweek from index
df["hour"] = df.index.hour
df["dayofweek"] = df.index.dayofweek

# One-hot encode hour (0â€“23) and dayofweek (0â€“6), convert to int
hour_dummies = pd.get_dummies(df["hour"], prefix="", prefix_sep="").astype(int)
day_dummies = pd.get_dummies(df["dayofweek"], prefix="", prefix_sep="").astype(int)
hour_dummies.index = df.index
day_dummies.index = df.index

# Combine features in notebook order (optionally include water lag)
X = pd.concat([
    hour_dummies,                 # 24 columns (0â€“23)
    day_dummies,                  # 7 columns (0â€“6)
    df[["water_m3_lag1"]],        # Optional extra feature (your enhancement)
    df[["temperature_C"]]         # Final feature
], axis=1)

# Target remains electricity
y = df["electricity_kW"]
```

```python
X.head()
```

```python
# --- STEP 2: Train/Test Split ---

X_train = X[X.index.month.isin(training_months)]
X_test = X[X.index.month.isin(test_months)]
y_train = y[y.index.month.isin(training_months)]
y_test = y[y.index.month.isin(test_months)]
```

```python
X_train.head()
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np
import pandas as pd

# --- Step 1: Define the hyperparameter grid ---
param_grid = {
    'n_estimators': [100],  # Keep fixed for speed; expand if needed
    'max_depth': [6, 8, 12],
    'min_samples_split': [2, 5, 10],
    'max_features': ['sqrt']
}

# --- Step 2: Initialize base model ---
base_model = RandomForestRegressor(random_state=42)

# --- Step 3: Setup GridSearchCV ---
grid_search = GridSearchCV(
    estimator=base_model,
    param_grid=param_grid,
    cv=3,                    # 3-fold cross-validation
    scoring='r2',
    n_jobs=-1,               # Use all cores
    verbose=2
)

# --- Step 4: Fit to training data ---
grid_search.fit(X_train, y_train)

# --- Step 5: Use best model for predictions ---
best_model = grid_search.best_estimator_
y_pred = best_model.predict(X_test)

# --- Step 6: Evaluate performance ---
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)
mape = 100 * np.mean(abs(y_pred - y_test) / y_test)

print(f"âœ… Best Parameters: {grid_search.best_params_}")
print(f"âœ… Test RMSE: {rmse:.2f}")
print(f"âœ… Test RÂ²: {r2:.3f}")
print(f"âœ… Test MAPE: {mape:.2f}%")

# Optional: Save results in DataFrame
predicted_vs_actual = pd.DataFrame({
    "Actual": y_test,
    "Predicted": y_pred
}, index=y_test.index)
```

```python
# Predict on training data
train_pred = best_model.predict(X_train)

# Evaluate on training set
train_rmse = np.sqrt(mean_squared_error(y_train, train_pred))
train_r2 = r2_score(y_train, train_pred)

print(f"Train RMSE: {train_rmse:.2f}")
print(f"Train RÂ²: {train_r2:.3f}")
```
